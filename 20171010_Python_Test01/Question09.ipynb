```python
#장르 추출 방법C
by_genre_generator = (set(x.split('|')) for x in data.genres) #여기서 제너레이터를 생성한다. 
#아마도 이 제너레이터는 data.genres 의 값들을 |를 구분자로 하여 split 한 후 생긴 3 값들을 순차적으로 읽어올 것이다. 
genres = sorted(set.union(*by_genre_generator))
print("유일한 장르 수 : ",len(genres))
genres
```

```python
np.zeros((len(data),len(genres))) #행렬생성 함수이다. 이러한 Array에 장르로 dummy한 결과를 담을 것이다. 
```

```python
dummies = DataFrame(np.zeros((len(data),len(genres))), columns = genres)
dummies.head(10)


#이것은 말 그대로 모두 0으로 채워진 아래와 같은 표만 만든 것이다. 
```

```python
#enumerate 의 이해 : index number와 해당 value 를 함께 반환해준다.

list1 = []
for i in range(10):
    list1.append(i*i)
print(*enumerate(list1))
```

```python
#enumerate 통해 dict 만들기
dict((i,v) for i,v in enumerate(list1))
```

```python
data.genres.head(10)#data.genres 라는 데이터는 pandas series 데이터이다.
```

```python
#실질적인 값을 채우는 단계 
#먼저 enumerate 를 통해 data.genres의 Series 데이터를 순차적으로, i, gen 에 담고 
#.ix 였으나 이것이 안중요하게(deprecated) 된다고 하니, loc를 쓰자. 
#위에서 생성한 dummies 변수(표)에 해당 row 의 genres 가 일치하면 1을 담는다. 단 gen은 split 한다. 

for i, gen in enumerate(data.genres):
    dummies.loc[i, gen.split('|')] = 1
```

```python
#여기서는 concat 의 기능으로 join을 사용한듯 하다. 
#기존의 영화 정보 데이터와 (data) (dummies)를 합친다. prefix는 붙이지 않기로 한다. 
movie_by_G = data.join(dummies)
movie_by_G
```

```python
#정렬 전 장르별로 정렬하기
movie_by_G[genres].sum()
# movie_by_G.sum()[genre]
```

```python
#장르별 합계를 구한다. 이것을 시리즈로 추출한다. 

result = Series(movie_by_G.sum()[genres])
result.sort_values()
```
