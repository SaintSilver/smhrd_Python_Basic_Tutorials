```python
import pandas as pd
train = pd.read_csv("train.csv", index_col="PassengerId")
print(train.shape)
train.head()
```

```python
import pandas as pd
test = pd.read_csv("test.csv", index_col="PassengerId")
print(test.shape)
test.head()
```

#preprocessing

```python
train.loc[train["Sex"]=="male", "Sex"] = 0
```

```python
print(train.shape)
```

```python
train.loc[train["Sex"]=="female", "Sex"] = 1
```

```python
train.head()
```

```python
test.loc[test["Sex"]=="male", "Sex"] = 0
test.loc[test["Sex"]=="female", "Sex"] = 1
```

```python
test[test["Fare"].isnull()]
```

```python
test.loc[test["Fare"].isnull(), "Fare"] = train["Fare"].mean()
test.loc[1044]
```

```python
train_embarked = pd.get_dummies(train["Embarked"], prefix="Embarked")
```

```python
test_embarked = pd.get_dummies(test["Embarked"], prefix="Embarked")
```

```python
train = pd.concat([train,train_embarked], axis = 1)
```

```python
test = pd.concat([test,test_embarked], axis = 1)
```

```python
train.head(10)
```

```python
test.head()
```

#Train.학습

```python
feature_names = ["Pclass","Sex", "Fare", "Embarked_C", "Embarked_Q", "Embarked_S"]
feature_names
```

```python
X_train = train[feature_names]
X_train.head()
```

```python
label_name = "Survived"
y_train = train[label_name]
y_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=5)
model
model.fit(X_train, y_train)
X_test = test[feature_names]
X_test.head()
```

```python
model.predict(X_test)
```

```python
predictions = model.predict(X_test)
predictions
```

```python
submission=pd.read_csv("gender_submission.csv", index_col="PassengerId")
submission["Survived"]=predictions
submission.head(100)
```

```python
submission.to_csv("decision-tree.csv")
```
