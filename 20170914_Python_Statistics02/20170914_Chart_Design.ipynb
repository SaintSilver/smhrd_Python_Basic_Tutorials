```python
%matplotlib nbagg

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

fig = plt.figure()
```

```python
ax = fig.add_subplot(1,1,1)
```

```python
ax.plot(np.random.rand(1000).cumsum()) 
#ax는 list 이다. 현재 줄의 처리결과를 ax 에 담아버리면 ax 는 더 이상 list 가 아니기 때문에 다음 줄부터 오류를 반환할것이다.
```

```python
ticks = ax.set_xticks([0,250,500,750,1000])
```

```python
labels = ax.set_xticklabels(["one","two","three","four","five"], rotation = 30, fontsize = "small")
```

```python
ax.set_title("Random walk plot")
```

```python
ax.set_xlabel("Stages")
ax.set_ylabel("Values") #철자에 주의하자, lable 아니고 label 이다.
```

```python
fig2= plt.figure()
```

```python
ax2 = fig2.add_subplot(1,1,1)
ax2.plot(np.random.randn(1000).cumsum()) 
ax2.plot(np.random.randn(1000).cumsum(),'k',label="one") 
ax2.plot(np.random.randn(1000).cumsum(),'b--',label="two") 
ax2.plot(np.random.randn(1000).cumsum(),'r',label="three") 
```

```python
ax2.legend(loc="best")
```
