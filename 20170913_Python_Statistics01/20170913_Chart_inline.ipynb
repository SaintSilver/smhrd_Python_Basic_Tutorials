차트를 그리기 위한 Importing

```python
%matplotlib inline

#위 코드는 수정을 할 수 있느 차트이고 
#수정을 원하지 않을 때 아래와 같이 옵션을 변경한다.
#%matplotlib in nbagg

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#임의의 시리즈 생성
s = pd.Series(np.random.rand(10).cumsum(), index=np.arange(0,100,10))
```

```python
#차트 표시(기본형 : 꺾은선형)
s.plot()
```

```python
#4개의 열을 가진 행렬 생성과 차트 표시
df2=pd.DataFrame(np.random.rand(10,4).cumsum(axis=0), 
                 columns=["a","b","c","d"], 
                 index = np.arange(0,100,10))
df2.plot()
```

```python
s2 = pd.Series(np.random.rand(16),index=list("abcdefghijklmnop"))
```

```python
#차트의 종류별로
s2.plot(kind="bar")
```

```python
s2.plot(kind="barh")
```

```python
#4개의 열을 가진 행렬 생성과 차트 표시
df3=pd.DataFrame(np.random.rand(6,4),
                 columns=pd.Index(["a","b","c","d"], name="Genius"), 
                 index=["one","two","three","four","five","six"])
df3.plot(kind = "bar")
```

```python
df3.plot(kind="bar", stacked = True)
```

# 피어슨 상관계수

```python
x1 = np.random.normal(1,1, size = (100,1))
x2 = np.random.normal(-2,4, size = (100,1))
X = np.concatenate((x1,x2), axis=1)
```

```python
df2 = pd.DataFrame(X,columns = ["x1","x2"])
```

```python
plt.scatter(df2["x1"],df2["x2"])
#이는 상관계수가 0에 가깝다. 
#plt는 임의로 생성한 pyplot 객체이다.
```
