```python
battles_by_name=battles.set_index(["name"])
battles.head(10)
```

```python
large_battle_mask=(battles_by_name["attacker_size"]+battles_by_name["defender_size"])>10000
```

```python
large_battle = battles_by_name.loc[large_battle_mask,["attacker_size","defender_size"]]
large_battle
```

```python
ax2 = large_battle.plot(kind ="bar", stacked=True)
```

```python
ax2.set_xticklabels(large_battle.index, rotation = 15, fontsize = "small")
```
